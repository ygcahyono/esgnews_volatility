```python
ind_mix_df = pd.DataFrame()

for _, asset in enumerate(train_df.Asset.unique()):
    
    subset_df = coverage_df[coverage_df.PermID == asset]
    
    name = subset_df.iloc[0,4]
    industry = subset_df.iloc[0,7]
    
    subs = pd.DataFrame({
        'Asset': asset,
        'Name': name,
        'Industry': industry,
        }, index= [_]
    )
    
    ind_mix_df = pd.concat([subs,ind_mix_df])
```
